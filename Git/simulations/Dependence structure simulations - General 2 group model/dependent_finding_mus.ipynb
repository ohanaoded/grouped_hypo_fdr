# 1.1.) FDR

```python
def func_OMTFDR(mu_FDR_variable, musPFDR, musMFDR, mu0, variance_0, mu1, variance_1, number_of_blocks, number_block_members, Rho, prob_to_1, maxiter, alpha):
    start = time.time()
    
    # locfdr generation (NOT DB)
    ############################################################ 1.) Generating locfdr & vec_h per group, as a List
    locfdr_unpacked, vec_h_unpacked = d_f.locfdr_generator_all_groups(mu0, variance_0, mu1, variance_1, number_of_blocks, number_block_members, Rho, prob_to_1, maxiter)
    ############################################################ 2.) packing locfdr & vec_h together for the rest 4 procedures 
    locfdr_packed = d_f.locfdr_concator(locfdr_unpacked)
    olocfdr_packed = np.sort(locfdr_packed, axis=1)
    
    ############################################################ 2.) FDEP ORACLE
    lev_mat_r, pow_mat_r, minprob_mat_r, ev_mat_r, er_mat_r = d_f.fdep_ORACLE([mu_FDR_variable], musPFDR, musMFDR, olocfdr_packed, maxiter, alpha)    
    
    print("This is MU now!@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@: ")
    print(mu_FDR_variable)
    stop = time.time()
    duration = stop-start
    print(f"Time this {maxiter} iterations run took: " + str(duration))
    print("This is the calculation right now, aiming for 0: ")
    print(lev_mat_r[0][0] - alpha)
    
    return(lev_mat_r[0][0] - alpha)
```

```python
%%time
maxiter = 5000

Rho = [0, 0.5]

root_OMTFDR = scipy.optimize.brentq(f=func_OMTFDR, a=430, b=490, args=(musPFDR, musMFDR, mu0, variance_0, mu1, variance_1, number_of_blocks, number_block_members, Rho, prob_to_1, maxiter, alpha)) #, maxiter=17, disp=False) 
root_OMTFDR
```

# 1.3.) mFDR 

```python
def func_OMTmFDR(mu_mFDR_variable, musFDR, musPFDR, mu0, variance_0, mu1, variance_1, number_of_blocks, number_block_members, Rho, prob_to_1, maxiter, alpha):
    start = time.time()
    
    # locfdr generation (NOT DB)
    ############################################################ 1.) Generating locfdr & vec_h per group, as a List
    locfdr_unpacked, vec_h_unpacked = d_f.locfdr_generator_all_groups(mu0, variance_0, mu1, variance_1, number_of_blocks, number_block_members, Rho, prob_to_1, maxiter)
    ############################################################ 2.) packing locfdr & vec_h together for the rest 4 procedures 
    locfdr_packed = d_f.locfdr_concator(locfdr_unpacked)
    olocfdr_packed = np.sort(locfdr_packed, axis=1)
    
    ############################################################ 2.) FDEP ORACLE
    lev_mat_r, pow_mat_r, minprob_mat_r, ev_mat_r, er_mat_r = d_f.fdep_ORACLE(musFDR, musPFDR, [mu_mFDR_variable], olocfdr_packed, maxiter, alpha)    

    print("This is MU now!@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@: ")
    print(mu_mFDR_variable)
    stop = time.time()
    duration = stop-start
    print(f"Time this {maxiter} iterations run took: " + str(duration))
    print("This is the calculation right now, aiming for 0: ")
    print(ev_mat_r[0][2] / er_mat_r[0][2] - alpha)

    return(ev_mat_r[0][2] / er_mat_r[0][2] - alpha)
```
